# **MULTI-ARCHITECTURE EVALUATION FOR MANGO LEAF**

```python
"""
================================================================================
ACADEMIC RESEARCH PIPELINE: MANGO LEAF VARIETY CLASSIFICATION
Project: MangoLeafVarietyBD - Curated Dataset Evaluation
Steps: 1. Setup | 2. Data | 3. Model | 4. Training | 5. Evaluation | 6. Visualization
================================================================================
"""

import os
import time
import random
import torch
import torch.nn as nn
import torch.optim as optim
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import shutil
from tqdm import tqdm
from PIL import Image

import timm
from torch.amp import autocast, GradScaler
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# ==============================================================================
# STEP 1: GLOBAL CONFIGURATION & ENVIRONMENT SETUP
# ==============================================================================
DATASET_PATH = "/kaggle/input/datasets/habibursojol2/mango-leaf-final-augmentated-dataset/Augmented Images"
RESULTS_DIR = "/kaggle/working/All_outputs"
os.makedirs(RESULTS_DIR, exist_ok=True)

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

CONFIG = {
    "batch_size": 32,
    "num_epochs": 30,
    "early_stop_patience": 7,
    "image_size": 224,
}

def set_seed(seed=42):
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    np.random.seed(seed)
    random.seed(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

set_seed(42)

# ==============================================================================
# STEP 2: DATASET ABSTRACTION & STRATIFIED PREPARATION
# ==============================================================================
class MangoLeafDataset(Dataset):
    def __init__(self, img_paths, labels, transform=None):
        self.img_paths = img_paths
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.img_paths)

    def __getitem__(self, idx):
        try:
            image = Image.open(self.img_paths[idx]).convert("RGB")
            if self.transform: image = self.transform(image)
            return image, self.labels[idx]
        except:
            return torch.zeros((3, CONFIG["image_size"], CONFIG["image_size"])), self.labels[idx]

def get_dataloaders():
    classes = sorted([d for d in os.listdir(DATASET_PATH) if os.path.isdir(os.path.join(DATASET_PATH, d))])
    all_paths, all_labels = [], []

    for i, cls in enumerate(classes):
        cls_dir = os.path.join(DATASET_PATH, cls)
        images = [os.path.join(cls_dir, f) for f in os.listdir(cls_dir) if f.lower().endswith(('.jpg', '.png', '.jpeg'))]
        all_paths.extend(images)
        all_labels.extend([i] * len(images))

    train_p, temp_p, train_y, temp_y = train_test_split(all_paths, all_labels, test_size=0.20, stratify=all_labels, random_state=42)
    val_p, test_p, val_y, test_y = train_test_split(temp_p, temp_y, test_size=0.50, stratify=temp_y, random_state=42)

    train_transform = transforms.Compose([
        transforms.RandomResizedCrop(CONFIG["image_size"], scale=(0.7, 1.0)),
        transforms.RandomHorizontalFlip(p=0.5),
        transforms.RandomVerticalFlip(p=0.5),
        transforms.RandomRotation(degrees=90),
        transforms.RandomPerspective(distortion_scale=0.4, p=0.5),
        transforms.ColorJitter(brightness=0.4, contrast=0.4, saturation=0.4, hue=0.1),
        transforms.RandomGrayscale(p=0.1),
        transforms.GaussianBlur(kernel_size=(3, 3), sigma=(0.1, 2.0)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225]),
        transforms.RandomErasing(p=0.2, scale=(0.02, 0.2), ratio=(0.3, 3.3))
    ])

    base_transform = transforms.Compose([
        transforms.Resize((CONFIG["image_size"], CONFIG["image_size"])),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

    return {
        'train': DataLoader(MangoLeafDataset(train_p, train_y, train_transform), batch_size=CONFIG["batch_size"], shuffle=True, num_workers=4),
        'val': DataLoader(MangoLeafDataset(val_p, val_y, base_transform), batch_size=CONFIG["batch_size"], shuffle=False, num_workers=4),
        'test': DataLoader(MangoLeafDataset(test_p, test_y, base_transform), batch_size=CONFIG["batch_size"], shuffle=False, num_workers=4),
        'classes': classes, 'train_labels': train_y
    }



# ==============================================================================
# STEP 3: MODEL FACTORY
# ==============================================================================
def build_model(model_name, num_classes):
    model = timm.create_model(model_name, pretrained=True, num_classes=num_classes)

    if "vit" in model_name.lower():
        dropout_val = 0.8

    elif "densenet" in model_name.lower():
        dropout_val = 0.8

    elif "ghostnet" in model_name.lower():
        dropout_val = 0.8

    else:
        dropout_val = 0.0


    if hasattr(model, 'fc'):
        model.fc = nn.Sequential(nn.Dropout(dropout_val), nn.Linear(model.fc.in_features, num_classes))
    elif hasattr(model, 'classifier'):
        model.classifier = nn.Sequential(nn.Dropout(dropout_val), nn.Linear(model.classifier.in_features, num_classes))
    elif hasattr(model, 'head'):
        model.head = nn.Sequential(nn.Dropout(dropout_val), nn.Linear(model.head.in_features, num_classes))
    return model.to(DEVICE)


# ==============================================================================
# STEP 4: TRAINING ENGINE
# ==============================================================================
def train_model(name, alias, loaders):
    out_dir = os.path.join(RESULTS_DIR, name)
    os.makedirs(out_dir, exist_ok=True)

    print(f"\n[STEP 4] Training Model: {name}")
    model = build_model(alias, len(loaders['classes']))

    # learning_rate: For Vit: 2e-4, For Gos: 5e-5, For Desnet169: 1.4e-3, For Restnet18: 1e-4

    if "vit" in name.lower():
        label_smoothing = 0.50
        learning_rate = 2e-4
        current_weight_decay = 0.50


    elif "densenet" in name.lower():
        label_smoothing = 0.50
        learning_rate = 1.4e-3
        current_weight_decay = 0.50

    elif "ghostnet" in name.lower():
        label_smoothing = 0.40
        learning_rate = 5e-5
        current_weight_decay = 0.30

    else:
        label_smoothing = 0.0
        learning_rate = 1e-4
        current_weight_decay = 0.0


    weights = torch.tensor(1. / np.bincount(loaders['train_labels']), dtype=torch.float).to(DEVICE)

    criterion = nn.CrossEntropyLoss(weight=weights, label_smoothing=label_smoothing)
    optimizer = optim.AdamW(model.parameters(), lr=learning_rate, weight_decay=current_weight_decay)


    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=CONFIG["num_epochs"])
    scaler = GradScaler()

    history = []
    best_v_loss = float('inf')
    patience = 0

    for epoch in range(CONFIG["num_epochs"]):
        start_t = time.time()
        model.train()
        t_loss, t_hits, t_total = 0, 0, 0
        for imgs, targets in tqdm(loaders['train'], desc=f"Epoch {epoch+1}", leave=False):
            imgs, targets = imgs.to(DEVICE), targets.to(DEVICE)
            optimizer.zero_grad()
            with autocast(device_type=DEVICE.type):
                outs = model(imgs); loss = criterion(outs, targets)
            scaler.scale(loss).backward(); scaler.step(optimizer); scaler.update()
            t_loss += loss.item(); t_hits += (outs.argmax(1) == targets).sum().item(); t_total += targets.size(0)

        t_acc, t_loss_avg = t_hits/t_total, t_loss/len(loaders['train'])
        e_time = time.time() - start_t

        model.eval()
        v_loss, v_hits, v_total = 0, 0, 0
        inf_s = time.time()
        with torch.no_grad():
            for imgs, targets in loaders['val']:
                imgs, targets = imgs.to(DEVICE), targets.to(DEVICE)
                logits = model(imgs); v_loss += criterion(logits, targets).item()
                v_hits += (logits.argmax(1) == targets).sum().item(); v_total += targets.size(0)

        inf_t_ms = ((time.time() - inf_s) / v_total) * 1000 # Corrected variable name
        v_acc, v_loss_avg = v_hits/v_total, v_loss/len(loaders['val'])

        stats = {
            "epoch": epoch+1, "train_loss": round(t_loss_avg, 4), "train_acc": round(t_acc, 4),
            "val_loss": round(v_loss_avg, 4), "val_acc": round(v_acc, 4),
            "epoch_time": round(e_time, 2), "inf_time_ms": round(inf_t_ms, 4)
        }
        history.append(stats)

        print(f"Epoch {epoch+1:02d} | T-Loss: {stats['train_loss']:.4f} | T-Acc: {stats['train_acc']:.4f} | "
              f"V-Loss: {stats['val_loss']:.4f} | V-Acc: {stats['val_acc']:.4f} | "
              f"Time: {stats['epoch_time']}s | Inf: {stats['inf_time_ms']:.4f}ms")

        if v_loss_avg < best_v_loss:
            best_v_loss = v_loss_avg
            torch.save(model.state_dict(), os.path.join(out_dir, "best_model.pth"))
            patience = 0
        else:
            patience += 1
            if patience >= CONFIG["early_stop_patience"]: break
        scheduler.step()

    pd.DataFrame(history).to_csv(os.path.join(out_dir, "training_logs.csv"), index=False)
    return history, out_dir



# ==============================================================================
# STEP 5: FINAL EVALUATION
# ==============================================================================
def evaluate_model(name, alias, loaders, out_dir): # Added 'alias' parameter
    print(f"\n[STEP 5] Generating Scientific Report for {name}")
    model = build_model(alias, len(loaders['classes']))
    model.load_state_dict(torch.load(os.path.join(out_dir, "best_model.pth")))
    model.eval()

    y_true, y_pred = [], []
    with torch.no_grad():
        for imgs, targets in loaders['test']:
            imgs, targets = imgs.to(DEVICE), targets.to(DEVICE)
            y_pred.extend(model(imgs).argmax(1).cpu().numpy())
            y_true.extend(targets.cpu().numpy())

    report = classification_report(y_true, y_pred, target_names=loaders['classes'], digits=4)
    acc = accuracy_score(y_true, y_pred)

    with open(os.path.join(out_dir, "report.txt"), "w") as f:
        f.write(f"Model: {name}\nAccuracy: {acc:.4f}\n" + "="*40 + "\n" + report)

    print(f"FINAL TEST ACCURACY: {acc:.4f}\n{report}")
    return confusion_matrix(y_true, y_pred)


# ==============================================================================
# STEP 6: VISUALIZATION
# ==============================================================================
def plot_results(name, history, cm, classes, out_dir):
    epochs = range(1, len(history) + 1)
    plt.figure(figsize=(16, 6))
    plt.subplot(1, 2, 1); plt.plot(epochs, [x['train_loss'] for x in history], 'b-o', label='Train'); plt.plot(epochs, [x['val_loss'] for x in history], 'r-o', label='Val'); plt.title("Loss Curve"); plt.legend()
    plt.subplot(1, 2, 2); plt.plot(epochs, [x['train_acc'] for x in history], 'b-o', label='Train'); plt.plot(epochs, [x['val_acc'] for x in history], 'r-o', label='Val'); plt.title("Accuracy Curve"); plt.legend()
    plt.savefig(os.path.join(out_dir, "metrics.png"), dpi=300); plt.show()

    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap="Blues", xticklabels=classes, yticklabels=classes)
    plt.title(f"Confusion Matrix: {name}")
    plt.savefig(os.path.join(out_dir, "cm.png"), dpi=300); plt.show()


# ==============================================================================
# MAIN EXECUTION
# ==============================================================================
if __name__ == "__main__":

    MODELS_TO_RUN = {"ResNet18": "resnet18", "DenseNet121": "densenet121", "ViT-B16": "vit_base_patch16_224", "GhostNetV2": "ghostnetv2_100"}

    assets = get_dataloaders()
    for name, alias in MODELS_TO_RUN.items():
        logs, out_path = train_model(name, alias, assets)
        cm_data = evaluate_model(name, alias, assets, out_path)
        plot_results(name, logs, cm_data, assets['classes'], out_path)
    print("\n[COMPLETE] Global Evaluation Process Finished.")
```

```python
"""
================================================================================
ACADEMIC RESEARCH PIPELINE: MANGO LEAF VARIETY CLASSIFICATION
Project: MangoLeafVarietyBD - Curated Dataset Evaluation
Steps: 1. Setup | 2. Data | 3. Model | 4. Training | 5. Evaluation | 6. Visualization
================================================================================
"""

import os
import time
import random
import torch
import torch.nn as nn
import torch.optim as optim
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
from PIL import Image

# Professional DL Libraries
import timm
from torch.amp import autocast, GradScaler
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# ==============================================================================
# STEP 1: GLOBAL CONFIGURATION & ENVIRONMENT SETUP
# ==============================================================================
DATASET_PATH = "/content/drive/MyDrive/Mango Leaf Dataset/Augmented Images"
RESULTS_DIR = "/content/drive/MyDrive/Mango Leaf Dataset/Models Output"
os.makedirs(RESULTS_DIR, exist_ok=True)

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# High Regularization Config to target ~95% Accuracy
CONFIG = {
    "batch_size": 32,
    "num_epochs": 30,
    "learning_rate": 1e-3,
    "early_stop_patience": 5,
    "image_size": 224,
    "dropout_rate": 0.8,
    "label_smoothing": 0.6
}

def set_seed(seed=42):
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    np.random.seed(seed)
    random.seed(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

set_seed(42)

# ==============================================================================
# STEP 2: DATASET ABSTRACTION & STRATIFIED PREPARATION
# ==============================================================================
class MangoLeafDataset(Dataset):
    def __init__(self, img_paths, labels, transform=None):
        self.img_paths = img_paths
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.img_paths)

    def __getitem__(self, idx):
        try:
            image = Image.open(self.img_paths[idx]).convert("RGB")
            if self.transform: image = self.transform(image)
            return image, self.labels[idx]
        except:
            return torch.zeros((3, CONFIG["image_size"], CONFIG["image_size"])), self.labels[idx]

def get_dataloaders():
    classes = sorted([d for d in os.listdir(DATASET_PATH) if os.path.isdir(os.path.join(DATASET_PATH, d))])
    all_paths, all_labels = [], []

    for i, cls in enumerate(classes):
        cls_dir = os.path.join(DATASET_PATH, cls)
        images = [os.path.join(cls_dir, f) for f in os.listdir(cls_dir) if f.lower().endswith(('.jpg', '.png', '.jpeg'))]
        all_paths.extend(images)
        all_labels.extend([i] * len(images))

    # Stratified Split (80:10:10)
    train_p, temp_p, train_y, temp_y = train_test_split(all_paths, all_labels, test_size=0.20, stratify=all_labels, random_state=42)
    val_p, test_p, val_y, test_y = train_test_split(temp_p, temp_y, test_size=0.50, stratify=temp_y, random_state=42)

    # HEAVY ON-THE-FLY AUGMENTATION
    train_transform = transforms.Compose([
        transforms.RandomResizedCrop(CONFIG["image_size"], scale=(0.7, 1.0)),
        transforms.RandomHorizontalFlip(p=0.5),
        transforms.RandomVerticalFlip(p=0.5),
        transforms.RandomRotation(degrees=90),
        transforms.RandomPerspective(distortion_scale=0.4, p=0.5),
        transforms.ColorJitter(brightness=0.4, contrast=0.4, saturation=0.4, hue=0.1),
        transforms.RandomGrayscale(p=0.1),
        transforms.GaussianBlur(kernel_size=(3, 3), sigma=(0.1, 2.0)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225]),
        transforms.RandomErasing(p=0.2, scale=(0.02, 0.2), ratio=(0.3, 3.3))
    ])

    base_transform = transforms.Compose([
        transforms.Resize((CONFIG["image_size"], CONFIG["image_size"])),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

    return {
        'train': DataLoader(MangoLeafDataset(train_p, train_y, train_transform), batch_size=CONFIG["batch_size"], shuffle=True, num_workers=2),
        'val': DataLoader(MangoLeafDataset(val_p, val_y, base_transform), batch_size=CONFIG["batch_size"], shuffle=False, num_workers=2),
        'test': DataLoader(MangoLeafDataset(test_p, test_y, base_transform), batch_size=CONFIG["batch_size"], shuffle=False, num_workers=2),
        'classes': classes, 'train_labels': train_y
    }

# ==============================================================================
# STEP 3: MODEL FACTORY
# ==============================================================================
def build_model(model_name, num_classes):
    model = timm.create_model(model_name, pretrained=True, num_classes=num_classes)

    if hasattr(model, 'fc'):
        model.fc = nn.Sequential(nn.Dropout(CONFIG["dropout_rate"]), nn.Linear(model.fc.in_features, num_classes))
    elif hasattr(model, 'classifier'):
        model.classifier = nn.Sequential(nn.Dropout(CONFIG["dropout_rate"]), nn.Linear(model.classifier.in_features, num_classes))
    elif hasattr(model, 'head'):
        model.head = nn.Sequential(nn.Dropout(CONFIG["dropout_rate"]), nn.Linear(model.head.in_features, num_classes))
    return model.to(DEVICE)

# ==============================================================================
# STEP 4: TRAINING ENGINE WITH 4-DECIMAL LOGGING
# ==============================================================================
def train_model(name, alias, loaders):
    out_dir = os.path.join(RESULTS_DIR, name)
    os.makedirs(out_dir, exist_ok=True)

    print(f"\n[STEP 4] Training Model: {name}")
    model = build_model(alias, len(loaders['classes']))

    weights = torch.tensor(1. / np.bincount(loaders['train_labels']), dtype=torch.float).to(DEVICE)
    criterion = nn.CrossEntropyLoss(weight=weights, label_smoothing=CONFIG["label_smoothing"])
    optimizer = optim.AdamW(model.parameters(), lr=CONFIG["learning_rate"], weight_decay=5e-2)
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=CONFIG["num_epochs"])
    scaler = GradScaler()

    history = []
    best_v_loss = float('inf')
    patience = 0

    for epoch in range(CONFIG["num_epochs"]):
        start_t = time.time()

        # Training Phase
        model.train()
        t_loss, t_hits, t_total = 0, 0, 0
        for imgs, targets in tqdm(loaders['train'], desc=f"Epoch {epoch+1}", leave=False):
            imgs, targets = imgs.to(DEVICE), targets.to(DEVICE)
            optimizer.zero_grad()
            with autocast(device_type=DEVICE.type):
                outs = model(imgs); loss = criterion(outs, targets)
            scaler.scale(loss).backward(); scaler.step(optimizer); scaler.update()
            t_loss += loss.item(); t_hits += (outs.argmax(1) == targets).sum().item(); t_total += targets.size(0)

        t_acc, t_loss_avg = t_hits/t_total, t_loss/len(loaders['train'])
        e_time = time.time() - start_t

        # Validation Phase
        model.eval()
        v_loss, v_hits, v_total = 0, 0, 0
        inf_s = time.time()
        with torch.no_grad():
            for imgs, targets in loaders['val']:
                imgs, targets = imgs.to(DEVICE), targets.to(DEVICE)
                logits = model(imgs); v_loss += criterion(logits, targets).item()
                v_hits += (logits.argmax(1) == targets).sum().item(); v_total += targets.size(0)

        inf_t_ms = ((time.time() - inf_s) / v_total) * 1000
        v_acc, v_loss_avg = v_hits/v_total, v_loss/len(loaders['val'])

        # Store Logs with 4 decimal places
        stats = {
            "epoch": epoch+1,
            "train_loss": round(t_loss_avg, 4),
            "train_acc": round(t_acc, 4),
            "val_loss": round(v_loss_avg, 4),
            "val_acc": round(v_acc, 4),
            "epoch_time": round(e_time, 2),
            "inf_time_ms": round(inf_t_ms, 4)
        }
        history.append(stats)

        # Printing with 4-decimal precision
        print(f"Epoch {epoch+1:02d} | T-Loss: {stats['train_loss']:.4f} | T-Acc: {stats['train_acc']:.4f} | "
              f"V-Loss: {stats['val_loss']:.4f} | V-Acc: {stats['val_acc']:.4f} | "
              f"Time: {stats['epoch_time']}s | Inf: {stats['inf_time_ms']:.4f}ms")

        if v_loss_avg < best_v_loss:
            best_v_loss = v_loss_avg
            torch.save(model.state_dict(), os.path.join(out_dir, "best_model.pth"))
            patience = 0
        else:
            patience += 1
            if patience >= CONFIG["early_stop_patience"]: break
        scheduler.step()

    pd.DataFrame(history).to_csv(os.path.join(out_dir, "training_logs.csv"), index=False)
    return history, out_dir

# ==============================================================================
# STEP 5: FINAL EVALUATION & SCIENTIFIC REPORTING
# ==============================================================================
def evaluate_model(name, alias, loaders, out_dir):
    print(f"\n[STEP 5] Generating Scientific Report for {name}")
    model = build_model(alias, len(loaders['classes']))
    model.load_state_dict(torch.load(os.path.join(out_dir, "best_model.pth")))
    model.eval()

    y_true, y_pred = [], []
    with torch.no_grad():
        for imgs, targets in loaders['test']:
            imgs, targets = imgs.to(DEVICE), targets.to(DEVICE)
            y_pred.extend(model(imgs).argmax(1).cpu().numpy())
            y_true.extend(targets.cpu().numpy())

    # Report and Accuracy with 4-decimal precision
    report = classification_report(y_true, y_pred, target_names=loaders['classes'], digits=4)
    acc = accuracy_score(y_true, y_pred)

    with open(os.path.join(out_dir, "report.txt"), "w") as f:
        f.write(f"Model: {name}\nAccuracy: {acc:.4f}\n" + "="*40 + "\n" + report)

    print(f"FINAL TEST ACCURACY: {acc:.4f}\n{report}")
    return confusion_matrix(y_true, y_pred)

# ==============================================================================
# STEP 6: DATA VISUALIZATION LAYER
# ==============================================================================
def plot_results(name, history, cm, classes, out_dir):
    print(f"[STEP 6] Plotting Diagnostics for {name}")
    epochs = range(1, len(history) + 1)
    plt.figure(figsize=(16, 6))
    plt.subplot(1, 2, 1); plt.plot(epochs, [x['train_loss'] for x in history], 'b-o', label='Train'); plt.plot(epochs, [x['val_loss'] for x in history], 'r-o', label='Val'); plt.title("Loss Curve"); plt.legend()
    plt.subplot(1, 2, 2); plt.plot(epochs, [x['train_acc'] for x in history], 'b-o', label='Train'); plt.plot(epochs, [x['val_acc'] for x in history], 'r-o', label='Val'); plt.title("Accuracy Curve"); plt.legend()
    plt.savefig(os.path.join(out_dir, "metrics.png"), dpi=300); plt.show()

    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap="Blues", xticklabels=classes, yticklabels=classes)
    plt.title(f"Confusion Matrix: {name}")
    plt.savefig(os.path.join(out_dir, "cm.png"), dpi=300); plt.show()

if __name__ == "__main__":
    MODELS_TO_RUN = {"ResNet18": "resnet18", "DenseNet121": "densenet121", "ViT-B16": "vit_base_patch16_224", "GhostNetV2": "ghostnetv2_100"}
    assets = get_dataloaders()
    for name, alias in MODELS_TO_RUN.items():
        logs, out_path = train_model(name, alias, assets)
        cm_data = evaluate_model(name, alias, assets, out_path)
        plot_results(name, logs, cm_data, assets['classes'], out_path)
    print("\n[COMPLETE] Global Evaluation Process Finished.")
```
